### 6. preprocessing and feature scaling:-

```python
## check the balance of data (target)
class_counts = data['output'].value_counts()
class_proportions = data['output'].value_counts(normalize=True)

# Count the number of occurrences of each class
print("Class counts:")
print(class_counts)

# Calculate class proportions
print("\nClass proportions:")
print(class_proportions)
```

``The dataset isnot balanced so,,,,,,,,,,,,,,,,,,``

```python
## split the data into Features (x) and Target (y)
x = data.drop('output', axis=1)
y = data['output']
```

```python
## split the data into train and test 
x_train , x_test , y_train , y_test = train_test_split(x, y, test_size=0.20, shuffle=True,
                                                       random_state=10)
## that mean 15 % of data for testing and 75 % for training
```

```python
## check shape
print('X_train shape -- ', x_train.shape)
print('y_train shape -- ', y_train.shape)
print('----------------------------------')
print('X_test shape  -- ' , x_test.shape)
print('y_test shape  -- ' , y_test.shape)
print('----------------------------------')
```

```python
## feature scaling 
slr = StandardScaler()
x_train = slr.fit_transform(x_train)
x_test  = slr.fit_transform(x_test)
```

``Now The dataset is ready for modeling``

### 7. Build models:-

```python
## Logistic Regression
logis_clr = LogisticRegression()
logis_clr.fit(x_train,y_train)

## predict and check accuracy
predicted = logis_clr.predict(x_test)
acc = accuracy_score(y_test, predicted)

print ("The accuracy of Logistic Regression is : ", acc*100)
print ("The recall   of Logistic Regression is : ", recall_score(y_test, predicted)*100)
```

```python
## K-nearest neighbor
knn_clr = KNeighborsClassifier(n_neighbors=8,p=1,metric='minkowski')
knn_clr.fit(x_train,y_train)

## predict and check accuracy
predicted = knn_clr.predict(x_test)
acc = accuracy_score(y_test, predicted)

print ("The accuracy of Logistic Regression is : ", acc*100)
print ("The recall   of Logistic Regression is : ", recall_score(y_test, predicted)*100)
```

```python
## Naive Bayes
NB_clr = GaussianNB()
NB_clr.fit(x_train,y_train)

## predict and check accuracy
predicted = NB_clr.predict(x_test)
acc = accuracy_score(y_test, predicted)

print ("The accuracy of Logistic Regression is : ", acc*100)
print ("The recall   of Logistic Regression is : ", recall_score(y_test, predicted)*100)
```

```python
## Support vector machine (polynomial kernel)
svm_clr = SVC(kernel='poly')

svm_clr.fit(x_train,y_train)

## predict and check accuracy
predicted = NB_clr.predict(x_test)
acc = accuracy_score(y_test, predicted)

print ("The accuracy of Logistic Regression is : ", acc*100)
print ("The recall   of Logistic Regression is : ", recall_score(y_test, predicted)*100)
```

```python
## use ensemble learning Voting Algorithm
voting_clr = VotingClassifier(voting='hard',n_jobs=-1,
                                            estimators=[
                                            ('logistic',logis_clr),
                                            ('KNN',knn_clr),
                                            ('Naive_bayes',NB_clr),
                                            ('SVM',svm_clr)
                                            ])
voting_clr.fit(x_train,y_train)

## predict and check accuracy
predicted = voting_clr.predict(x_test)
acc = accuracy_score(y_test, predicted)

print ("The accuracy of Logistic Regression is : ", acc*100)
print ("The recall   of Logistic Regression is : ", recall_score(y_test, predicted)*100)
```

```python
##  Extreme Gredient Boost Classifier
xgb_clr = xgb.XGBClassifier(n_estimators=100)
                            

xgb_clr.fit(x_train, y_train) 

## predict and check accuracy
ypred = xgb_clr.predict(x_test)
ac  = accuracy_score(y_test, ypred)
print ("The accuracy of Logistic Regression is : ",ac*100)
print ("The recall   of Logistic Regression is : ", recall_score(y_test, ypred)*100)
```

```python
## Random forest
RF_clr = RandomForestClassifier(n_estimators=11)

RF_clr.fit(x_train, y_train)  


## predict and check accuracy
ypred = RF_clr.predict(x_test)
ac = accuracy_score(y_test, ypred)
print ("The accuracy of Logistic Regression is : ", ac*100)
print ("The recall   of Logistic Regression is : ", recall_score(y_test, ypred)*100)
```
